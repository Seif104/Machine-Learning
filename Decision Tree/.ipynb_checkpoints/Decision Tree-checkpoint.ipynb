```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('drug200.csv',delimiter=',')
print(dataset.shape)
dataset.head(10)
```

```python
dataset['Drug'].value_counts() * 100 / dataset['Drug'].describe()[0]
```

```python
dataset['Cholesterol'].value_counts()
```

```python
x = dataset[['Age','Sex','BP','Cholesterol','Na_to_K']].values
x
```

```python
le = LabelEncoder()
x[:,1] = le.fit_transform(x[:,1])

x[:,2] = le.fit_transform(x[:,2])

x[:,3] = le.fit_transform(x[:,3])

x
```

```python
y = dataset['Drug'].values
y[:5]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 1)
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
decision_tree = tree.DecisionTreeClassifier(criterion = 'entropy').fit(x_train,y_train)
decision_tree
```

```python
predictions = decision_tree.predict(x_test)

print("Accuracy = ",accuracy_score(y_test,predictions))
```

```python
tree.plot_tree(decision_tree)
plt.show()
```

```python
decision_tree.predict_proba([[25,0,0,1,30.2]])
```

```python
 import graphviz
    
data = tree.export_graphviz(decision_tree,feature_names = ['Age','Sex','BP','Cholesterol','Na_to_K'],
                            class_names =['drugA','drugB','drugC','drugX','drugY'],rounded = True)
graph = graphviz.Source(data)
#graph.render("iriss")
graph
```
